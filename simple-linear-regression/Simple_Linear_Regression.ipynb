# **Simple Linear Regression**



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("homeprices.csv")
df.tail(3)
```

```python
plt.scatter(df.area,df.price,color='red',marker='+')
plt.xlabel('Area')
plt.ylabel('Price')
```

```python
#creating linear regression object 'reg'
reg=linear_model.LinearRegression()
reg.fit(df[['area']],df.price)
```

```python
reg.predict([[3200]])
```

```python
#to get values of m and b in y=mx+b
print(f"Value of m(Coefficient): {reg.coef_[0]}, Value of b (Intercept) : {reg.intercept_}")
```

```python
d = pd.read_csv("areas.csv")
d
```

```python
#Predicting prices of all areas
p = reg.predict(d)
p
```

```python
#Creating new column in dataframe d and filling it with prices in p
d['prices'] = p
d.head(3)
```

```python
d.to_csv("prediciton.csv",index=False)
```

```python
#Plotting the predicted line
plt.scatter(df.area,df.price,color='red',marker='+')
plt.xlabel('Area',fontsize=20)
plt.ylabel('Price',fontsize=20)
plt.plot(df.area,reg.predict(df[['area']]),color='blue')
```
